```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import metrics 
import sys
import os
import time
import concurrent.futures
from functools import partial
from subprocess import check_output

os.chdir('..')

models = ['m3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11']
kfolds = ['1', '2', '3', '4', '5']
```

```python
# variance in k folds
for m in models:
    mk_stats = []
    for k in kfolds:
        df = pd.read_csv(f'./prediction_data/{m}-k{k}.csv')
        
        confusion_matrix = metrics.confusion_matrix(
            df['Label'],
            df['Prediction'],
            normalize='true',
        )
        
        cm_display = metrics.ConfusionMatrixDisplay(
            confusion_matrix = confusion_matrix, 
            display_labels = [True, False]
        )

        cm_display.plot(cmap='Blues')
        plt.title(f'{m} k-fold{k} Confusion Matrix')
        plt.show() 
        
        mk_stats.append(confusion_matrix)

    print('true positive standard deviation: ' +
        '{:e}'.format(np.std([
            mk_stats[0][0][0], 
            mk_stats[1][0][0], 
            mk_stats[2][0][0], 
            mk_stats[3][0][0]
        ]))
    )
    
    print('false negative standard deviation: ' +
        '{:e}'.format(np.std([
            mk_stats[0][0][1], 
            mk_stats[1][0][1], 
            mk_stats[2][0][1], 
            mk_stats[3][0][1]
        ]))
    )

    print('false positive standard deviation: ' + 
        '{:e}'.format(np.std([
            mk_stats[0][1][0], 
            mk_stats[1][1][0], 
            mk_stats[2][1][0], 
            mk_stats[3][1][0]
        ]))
    )

    print('true negative standard deviation: ' +
        '{:e}'.format(np.std([
            mk_stats[0][1][1], 
            mk_stats[1][1][1], 
            mk_stats[2][1][1], 
            mk_stats[3][1][1]
        ]))
    )
    
    print(mk_stats)
```
