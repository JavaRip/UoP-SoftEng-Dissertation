```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import metrics 
import os
os.chdir('..')

from models.model_utils.evaluator import gen_eval

m3 = pd.read_csv('./prediction_data/m3-k1.csv')
m4 = pd.read_csv('./prediction_data/m4-k1.csv')
m5 = pd.read_csv('./prediction_data/m5-k1.csv')
m6 = pd.read_csv('./prediction_data/m6-k1.csv')
m7 = pd.read_csv('./prediction_data/m7-k1.csv')
m8 = pd.read_csv('./prediction_data/m8-k1.csv')
m9 = pd.read_csv('./prediction_data/m9-k1.csv')
m10 = pd.read_csv('./prediction_data/m10-k1.csv')
m11 = pd.read_csv('./prediction_data/m11-k1.csv')

nc_src = './well_data/src_data.csv'
df = pd.read_csv(nc_src)
```

```python
def get_eval(predictions, m_name, k_fold, upa):
    eval = gen_eval(predictions)
    eval_df = pd.DataFrame(data={
      'upa': upa,
      'model': m_name,
      'k': k_fold,
      'sensitivity': eval['sensitivity'],
      'f1': eval['f1'],
      'accuracy': eval['accuracy'],
      'precision': eval['precision'],
      'specificity': eval['specificity'],
    }, index=[0])

    return eval_df 

evals = []
num_upa = len(df['Upazila'].unique())
for upa in df['Upazila'].unique():
    upa_ind = df['Upazila'].unique().tolist().index(upa)
    print(f'upazila {upa_ind} / {num_upa}')
    
    evals.append(
        {
            'm3': get_eval(m3[m3['Upazila'] == upa], 'm3', 1, upa),
            'm4': get_eval(m4[m4['Upazila'] == upa], 'm4', 1, upa),
            'm5': get_eval(m5[m5['Upazila'] == upa], 'm5', 1, upa),
            'm6': get_eval(m6[m6['Upazila'] == upa], 'm6', 1, upa),
            'm7': get_eval(m7[m7['Upazila'] == upa], 'm7', 1, upa),
            'm8': get_eval(m8[m8['Upazila'] == upa], 'm8', 1, upa),
            'm9': get_eval(m9[m9['Upazila'] == upa], 'm9', 1, upa),
            'm10': get_eval(m10[m10['Upazila'] == upa], 'm10', 1, upa),
            'm11': get_eval(m11[m11['Upazila'] == upa], 'm11', 1, upa),
        }
    )
```

```python
print(evals[0]['m3'])
```

```python
for m in ['m3','m4','m5','m6','m7','m8','m9','m10','m11']:
    wins = []
    for e in evals:
        if np.greater(e[m]['accuracy'], e['m5']['accuracy'])[0]:
            if np.greater(e[m]['f1'], e['m5']['f1'])[0]:
                upa = e[m]['upa']
                wins.append({
                    'model': m,
                    'upa': upa[0],
                    'model_accuracy': e[m]['accuracy'][0],
                    'model_f1': e[m]['f1'][0],
                    'm5_accuracy': e['m5']['accuracy'][0],
                    'm5_f1': e['m5']['f1'][0],
                })
                

    print(f'{m} beats model 5 {len(wins)} times')
    print(wins)
    print('========================')
```
